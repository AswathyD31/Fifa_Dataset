# 3.Which player has the highest salary? 

```python
# Convert 'Wage' column to numeric format
fifa['Wage'] = fifa['Wage'].str.replace('€', '').str.replace('K', '').astype(int)
```

```python
# Make a new dataset named player_salary of short_name and wage_eur columns.
player_salary=fifa[['Name','Wage']]
player_salary
```

```python
# Head of the dataset.
player_salary.head()
```

```python
# Sort the dataset in the descending order of their salary.
player_salary=player_salary.sort_values(by='Wage',ascending=False)
player_salary
```

```python
# Top 5 records of the dataset.
player_salary.head()
```

```python
# Names of the top 5 players with highest salary.
list(player_salary['Name'][0:5])
```

```python
# Top 5 highest salary.
list(player_salary['Wage'][0:5])
```

##### # Data Story: Knowing the player with the highest salary provides insight into the commercial value and marketability of players. It reflects not only their on-field talent but also their off-field influence and brand value

# Plot for higest Salary player

```python
# Bar-plot of top 5 players with highest salary.
plt.figure(figsize=(6,5))
sns.barplot(x=list(player_salary['Name'][0:5]),y=list(player_salary['Wage'][0:5]),ec='black',lw=1)
plt.title("FIFA Salary",fontsize=15)
plt.xlabel("Name of the Players",fontsize=5)
plt.ylabel("Wage",fontsize=13)
plt.grid(True)
plt.show()
```

#  4.Plot a histogram to get the salary range of the players.

# 7.Which foot is most preferred by the players?Draw a bar chart for preferred foot? 

```python
# 7. Which foot is most preferred by the players?
fifa.value_counts(["Preferred Foot"])
```

# The most preferred foot by the players is the "Right foot".

```python
x=fifa.value_counts(["Preferred Foot"]).index
y=fifa.value_counts(["Preferred Foot"]).values
x1=list(x[0])
x2=list(x[1])
x=x1+x2
plt.figure(figsize=(5,10),dpi=100)
plt.bar(x,y,label="No of Players",color='red')
plt.title('MOST PREFERRED FOOT BY THE PLAYERS')
plt.yticks(range(0,15000,1000))# y axis numbers
plt.xlabel('PREFERRED FOOT')
plt.ylabel('No Of Players')
plt.legend() # shows label
plt.show()
```

##### # Data Story: The preferred foot analysis helps in understanding the distribution of left-footed versus right-footed players. This can be crucial for team formations, strategies, and player roles on the field, highlighting the need for a balance between left and right-footed players.
